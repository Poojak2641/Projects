```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df=pd.read_csv("Downloads/analysis_dataset.csv")
```

```python
df
```

```python
df['Category']=df['Category'].replace('Accessories','Electronics')
df
```

```python
#A. Basic Python (Data Types, If–Else, Loops)
#Print how many customers are above age 40.
count =0
for age in df['Age']:
    if age > 40:
        count +=1
print(f"{count} customers are above age 40")
```

```python
#Create a function that takes Price & Quantity and returns the total bill amount. Use this for the first 10 rows.
def Total_bill(price,quantity):
    return price * quantity

for i in range(10):
    price =df.loc[i,'Price']
    quantity=df.loc[i,'Quantity']
    print(f"Row {i+1}: Total Bill ",Total_bill(price,quantity))
```

```python
#Write a program to count how many customers purchased “Mobile” using loops (without pandas).
count = 0
for Mobile in df['Product']:
    if Mobile == 'Mobile':
        count +=1
print(f"{count} customers purchased 'Mobile'")
```

```python
#Find the longest name in the dataset using Python loops and string operations.
long_name=""
max_length=0

for name in df['Name']:
    if len(name) > max_length:
        long_name= name
        max_length=len(name)
print("Longest Name: ",long_name)
print("Maximum length: ",max_length)
```

```python
#Convert a user-given age to integer and handle errors with exception handling.
try:
    user_age=int(input("Enter the Age:"))
    
except ValueError:
    print("Invalid Input, Please Enter the correct number")
else:
    print("Age is ",user_age)
```

```python
#Accept a product from user input and print all customers who bought that product.
product_name =input("Enter a product Name: ")
for i in range(len(df)):
    if df.loc[i,'Product'].lower()== product_name.lower():
        print(df.loc[i,'CustomerID']," , ",df.loc[i,'Name']," , ",df.loc[i,'Product'])
```

```python
#Write a program to calculate average rating using only loops (no numpy/pandas).
total=0
count=0
for rating in df['Rating']:
    total += rating
    count +=1

average=total / count
print("Average Rating is, ",average)
    
```

```python
#B. Pandas (Cleaning + Analysis)
#Load the CSV file and check:
#Shape
print("Shape of dataset:",df.shape)

#Missing values
print("\nMissing values:\n",df.isnull().sum())

# Duplicates
print("\nDuplicate rows:", df.duplicated().sum())
```

```python
#Remove duplicates if any.
df.drop_duplicates().reset_index(drop=True)
print("Duplicates removed")
```

```python
#Handle missing values:
#Fill age with mean age
df['Age'].fillna(df['Age'].mean())
df['Age']
```

```python
#Fill rating with median rating
df['Rating'].fillna(df['Rating'].median())
```

```python
#Convert column PurchaseDate into datetime.
df["PurchaseDate"]=pd.to_datetime(df["PurchaseDate"])
df["PurchaseDate"]
```

```python
#Add a new column TotalAmount = Price × Quantity.
df['TotalAmount']=df['Price'] * df['Quantity']
df
```

```python
#Find top 5 highest-paying customers.
top_5= df.nlargest(5,'TotalAmount')[['CustomerID','Name','TotalAmount']]
top_5
```

```python
#Find which city has the highest number of purchases.
city_highest = df["City"].value_counts().idxmax()
count_purchases = df["City"].value_counts().max()

print("City with highest purchases:", city_highest)
print("Number of purchases:", count_purchases)
```

```python
#Group by Product and calculate:
#Total Quantity
#Average Price
#Average Rating
product = df.groupby("Product").agg(
    Total_Quantity=("Quantity", "sum"),
    Average_Price=("Price", "mean"),
    Average_Rating=("Rating", "mean")
)

product
```

```python
#Filter all purchases made using UPI.
UPI_purchases= df[df['PaymentMethod']=='UPI']
UPI_purchases.head()
```

```python
#Which category generated the highest revenue?
category_revenue = df.groupby('Category')['TotalAmount'].sum()
highest_category=category_revenue.idxmax()
amount = category_revenue.max()

print("highest category is :",highest_category)
print("Amount",amount)
```

```python
#C. NumPy Concepts
'''Convert Age column to a NumPy array and calculate:
mean
median
standard deviation'''

age=np.array(df['Age'])


print("Mean of Age",np.mean(age))
print("\nMedian of Age",np.median(age))
print("\nStandaed deviation of Age",np.std(age))

```

```python
#Normalize the Price column using NumPy (min-max scaling).
price=df['Price'].to_numpy()
max_price=price.max()
min_price=price.min()

Normalize_price= (price - min_price) / (max_price - min_price)
print("Normalize price",Normalize_price[:10])
```

```python
#Count how many values in Rating NumPy array are ≥ 4.
Rating=df['Rating'].to_numpy()
np.sum(Rating >=4)
```

```python
#D. Matplotlib & Seaborn (Visualizations)
#Plot a bar chart of product-wise sales quantity.
product_sales=df.groupby("Product")["Quantity"].sum()

plt.figure(figsize=(10,6))
plt.bar(product_sales.index, product_sales.values)
plt.xlabel("Product")
plt.ylabel("Total Quantity")
plt.title("Product-wise Sales Quantity")
plt.show()
```

```python
#Plot histogram of Age distribution.
plt.figure(figsize=(10,6))
sns.histplot(df['Age'],color='y',kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
#older or younger customers spend more
df['TotalSales'] =df['Price'] * df['Quantity']
plt.figure(figsize=(10,6))
sns.lineplot(x='Age',y='TotalSales',data=df,c='g')
plt.title("Age vs Spending Behaviour")
plt.xlabel("Customer Age")
plt.ylabel("Total Sales")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x="Gender", y='TotalSales',data=df)
plt.title("Male vs Female Spending")
plt.xlabel("Gender")
plt.ylabel("Total Sales")
plt.show()
```

```python
#Create a Seaborn countplot for PaymentMethod.
plt.figure(figsize=(10,6))
sns.countplot(x="PaymentMethod",data =df,hue='PaymentMethod')
plt.show()
```

```python
#Create a Seaborn boxplot for Price by Category.
plt.figure(figsize=(10,6))
sns.boxplot(x="Category",y="Price",data=df,color='g')
plt.title("Price Distribution by Category")
plt.show()
```
